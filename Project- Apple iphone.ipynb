```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv(r"F:\1. Data Cleaning\each_match_records.csv")
```

# Load Dataset

```python
dataset.head(5)
```

# Number of Matches Won by Each Team in IPL 2023

```python
winner_counts = dataset['winner'].value_counts()
plt.figure(figsize=(10,6))
winner_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Matches Won by Each Team in IPL 2023')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Top 3 Venues for Matches in IPL 2023

```python
venue_counts = dataset['venue'].value_counts()
top_venues = venue_counts[:3]

plt.figure(figsize=(6, 6))
plt.pie(top_venues, labels=top_venues.index, autopct='%0.2f%%', startangle=80, colors=["Red","Green","orange"])
plt.title('Top 3 Venues for Matches in IPL 2023', fontsize=16)
plt.legend(loc=1)
plt.savefig("Image5.jpg")
plt.show()
```

# Number of Matches Won by Punjab Kings at Each Location

```python
gt_wins = dataset[dataset['winner'] == 'Mumbai Indians']
win_counts = gt_wins['location'].value_counts().reset_index()
win_counts.columns = ['location', 'win_count']
plt.figure(figsize=(10, 5))
plt.bar(win_counts['location'], win_counts['win_count'], color='skyblue')
plt.title('Number of Matches Won by Punjab Kings at Each Location')
plt.xlabel('Location')
plt.ylabel('Number of Wins')
plt.xticks(rotation=5, ha='right')
plt.savefig("Number of Matches Won by Punjab Kings at Each Location.jpg") 
plt.show()
```

# Top 6 Winning Teams in IPL 2023

```python
plt.figure(figsize=(10,5))
dataset['winner'].value_counts()[:6]
x=dataset['winner'].value_counts()[:6].index
y=dataset['winner'].value_counts()[:6].values
plt.bar(x,y,label="Number of wins",color="skyblue")
plt.title('Top 6 Winning Teams in IPL 2023',fontsize=20)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Number of Wins',fontsize=15)
plt.xticks(rotation=20)
plt.legend()
plt.savefig("Image1.jpg")
plt.show()
```

# Number of Matches Played as Team 1 - IPL 2023

```python
team1_counts = dataset['team1'].value_counts()
plt.figure(figsize=(12, 7))
team1_counts.plot(kind='bar',   color ='skyblue')   #color='#E3242B')
plt.title('Number of Matches Played as Team 1 - IPL 2023', fontsize=20)
plt.xlabel('Teams', fontsize=16)
plt.ylabel('Number of Matches', fontsize=16)
plt.yticks([i for i in range(0,12,2)])
plt.xticks(rotation=45)
plt.legend(['Matches Played'], fontsize=16)
plt.savefig("Image2.jpg")
plt.show()
```

# How teams proceeded after winning the toss

```python
# (6) How teams proceeded after winning the toss
plt.figure(figsize=(4, 4))
x = dataset.groupby('toss_decision').size()
x.plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Toss')
plt.legend(loc=2)
plt.savefig("toss.jpg")
plt.show()
```
